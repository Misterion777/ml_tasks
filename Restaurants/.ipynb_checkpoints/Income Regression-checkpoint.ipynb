```python
new_df
```

## Prepare train test

```python
X = new_df.drop('Weekly_Sales',axis=1)
y = new_df['Weekly_Sales']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X[['Temperature','Fuel_Price','Unemployment']] = scaler.fit_transform(X[['Temperature','Fuel_Price','Unemployment']])
```

```python
X
```

## Models

```python
from sklearn.metrics import SCORERS
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import cross_val_score
```

```python
sorted(SCORERS.keys())
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor(random_state=42)
```

```python
scores = cross_val_score(reg, X,y, cv=5,scoring='neg_mean_squared_error',verbose=2)
```

```python
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print(scores)
    print(f'Mean: {scores.mean()}')
    print(f'Std: {scores.std()}')
```

```python
display_scores(tree_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(random_state=42)
```

```python
scores = cross_val_score(forest, X,y, cv=5,scoring='neg_mean_squared_error',verbose=2)
```

```python
forest_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(forest_rmse_scores)
```

```python
from sklearn.svm import SVR
```

```python
svm = SVR(kernel='linear')
```

```python
scores = cross_val_score(svm, X,y, cv=5,scoring='neg_mean_squared_error',verbose=2)
```

```python
svm_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(svm_rmse_scores)
```
